```python
def extract_pdf_text(data_folder, filename):
    pdf_path = os.path.join(data_folder, filename)
    all_text = []

    with pdfplumber.open(pdf_path) as pdf:
        for page in pdf.pages:
            filtered_page = page
            chars = filtered_page.chars

            for table in page.find_tables():
                first_table_character = page.crop(table.bbox).chars[0]
                filtered_page = filtered_page.filter(
                    lambda obj: get_bbox_overlap(obj_to_bbox(obj), table.bbox) is None
                )
                chars = filtered_page.chars

                # Extract table as a list of lists
                table_data = table.extract()

                # Convert table data into a DataFrame
                df = pd.DataFrame(table_data)
                df = df.fillna("")

                # Convert DataFrame to Markdown
                markdown = df.to_markdown(index=False)

                chars.append({**first_table_character, "text": markdown})

            page_text = extract_text(chars, layout=True)
            all_text.append(page_text)

    return "\n".join(all_text)
```

```python
filename = "1. MSA Gateware (Enteron) 1.pdf"
extracted_text = extract_pdf_text(data_folder, filename)
```
